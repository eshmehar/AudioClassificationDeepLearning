```python
def features_extractor(file):
    audio, sample_rate = librosa.load(file, res_type='kaiser_fast')
    mfccs_features = librosa.feature.mfcc(y=audio, sr=sample_rate, n_mfcc=40)
    mfccs_scaled_features = np.mean(mfccs_features.T, axis=0)
    return mfccs_scaled_features

def predict_audio_class(file):
    audio_features = features_extractor(file)
    audio_features = audio_features.reshape(1, 40, 1, 1)  # Reshape for the model
    prediction = model.predict(audio_features)
    predicted_class = labelencoder.inverse_transform(np.argmax(prediction, axis=1))
    return predicted_class

# Test with a new audio file
predicted_class = predict_audio_class('street.wav')
print(f'Predicted class: {predicted_class}')
```
